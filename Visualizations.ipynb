```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv('Resources/diabetes_012_health_indicators_BRFSS2015.csv') 

# Calculate the counts for each unique value in the "PhysHealth" column
physhealth_counts = data['PhysHlth'].value_counts().sort_index()

# Calculate the counts for each unique value in the "Diabetes_012" column
diabetes_counts = data['Diabetes_012'].value_counts().sort_index()

# Create a figure with two subplots (side by side)
plt.figure(figsize=(12, 5))  

# Plot the first bar chart for "PhysHealth" column
plt.subplot(1, 2, 1)
plt.bar(physhealth_counts.index, physhealth_counts.values, color='skyblue')
plt.xlabel('Physical Health')
plt.ylabel('Count')
plt.title('Distribution of Physical Health')

# Plot the second bar chart for "Diabetes_012" column
plt.subplot(1, 2, 2)
plt.bar(diabetes_counts.index, diabetes_counts.values, color='lightcoral')
plt.xlabel('Diabetes')
plt.ylabel('Count')
plt.title('Distribution of Diabetes')

plt.tight_layout() 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv('Resources/diabetes_012_health_indicators_BRFSS2015.csv')  

# Create a cross-tabulation for HighBP vs. Diabetes_012
cross_tab = pd.crosstab(data['HighBP'], data['Diabetes_012'], normalize='index')

# Create the grouped bar chart
cross_tab.plot(kind='bar', stacked=False, colormap='Set3', edgecolor='black', figsize=(8, 6))

plt.xlabel('High Blood Pressure')
plt.ylabel('Proportion')
plt.title('High Blood Pressure vs. Diabetes')
plt.xticks([0, 1], ['No High BP', 'High BP'], rotation=0)

plt.legend(title='Diabetes_012', loc='upper right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv('Resources/diabetes_012_health_indicators_BRFSS2015.csv')

# Define colors for the box plots
box_colors = ['lightblue', 'lightgreen', 'lightcoral']
labels = ['0', '1', '2']
data_to_plot = [data[data['Diabetes_012'] == i]['PhysHlth'] for i in range(3)]

# Create a box plot for each category
plt.figure(figsize=(8, 6))
bp = plt.boxplot(data_to_plot, labels=labels, patch_artist=True)

# Customize the box colors
for patch, color in zip(bp['boxes'], box_colors):
    patch.set_facecolor(color)

# Customize other plot elements
plt.xlabel('Diabetes')
plt.ylabel('Physical Health')
plt.title('Relationship Between Diabetes and Physical Health)')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add a legend
legend_labels = ['0 - No Diabetes', '1 - Pre-Diabetes', '2 - Diabetes']
plt.legend(legend_labels, loc='upper right', fontsize=10)

plt.show()
```
